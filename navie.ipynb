```
data = pd.read_csv('/content/drive/My Drive/ĐỒ ÁN MÁY HỌC/tieuduong.csv')
data.head()
```

```
#bảng tóm tắt số liệu thống kê liên quan đến các dữ liệu số liệu thống kê liên quan đến các dữ liệu
data.describe()
```

```
#Chia du lieu
from sklearn.model_selection import train_test_split
#seperate colums in dependent and independent variables(cột riêng trong các biến phụ thuộc và độc lập)
X=data.drop("Outcome",axis=1)
Y=data[["Outcome"]]
```

```
#split dataset into trainning and test sets(chia dữ liệu thành tập huấn luyện và kiểm tra)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```
#Huấn luyện và thực hiện dự đoán trên bộ dữ liệu thử
#Import gaussian Naive Bayes model(Mô hình Navie Bayes)
from sklearn.naive_bayes import GaussianNB
#Creat a Gaussian Classifier (Phân loại Gaussian)
model=GaussianNB()
#Train the model using the training sets
model.fit(X_train,Y_train.values.ravel())
#Predict the response for test datasets(Dự đoán đáp ứng cho bộ dữ liệu thử nghiệm)
Y_pred = model.predict(X_test)
```

```
#kiểm tra tần suất 
from sklearn import metrics
print("accuracy: ",metrics.accuracy_score (Y_test,Y_pred))
```
